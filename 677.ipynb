```python
import pandas as pd
import numpy
```

```python
# 初期値の設定
mu = 0.15  # 期待リターン
sigma = 0.3   # リスク（標準偏差）
dt = 0.004   # 時間間隔
S_0 = 35  # 初期投資額
```

```python
# data/rand1.csvを読み込む
df = pd.read_csv('data/rand1.csv')
```

```python
# exp(mu*dt + sd*sqrt(dt)*rand)を計算し初期値を掛ける  
df['V'] =  S_0 * ((mu * dt + sigma * (dt ** 0.5) * df['rand']).apply(lambda x: 2.718281828459045 ** x).cumprod())
```

$$
    S_{t+\Delta t} = S_t * \exp(\mu\Delta t + \sigma Z \sqrt{\Delta t})
$$

```python
# 株価の推移をmatplotlibで描画
import matplotlib.pyplot as plt
plt.plot(df['V'])
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.title('Stock Price Simulation')
plt.show()
```
